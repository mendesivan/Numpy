```python
import numpy as np
from numpy import random
sample = np.arange(8);print("Sample: ",sample); print("\n")
print("Sqrt of Sample:\n",np.sqrt(sample)); print("\n"); print("Exponential or Sample:\n",np.exp(sample))
```

```python
#Binary Functions take two values and others such as max, add, min, etc.
x = random.randn(3)
y = random.randn(3)

print(x); print("\n")
print(y); print("\n")
print(np.maximum(x,y)); print("\n")
print(np.modf(x)) #function modf returns the fractional and integral parts of a floating point array.
```
